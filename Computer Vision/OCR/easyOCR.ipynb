```python
import easyocr
import cv2
import matplotlib.pyplot as plt
```

```python
reader = easyocr.Reader(['en'], gpu=False)  # GPU memory is not enough
result = reader.readtext('cv.jpg')
result
```

```python
img = cv2.imread('cv.jpg')
for detection in result: 
    top_left = detection[0][0]
    top_left = (int(top_left[0]), int(top_left[1]))  # To be ensure that the coordinates are integers
    bottom_right = detection[0][2]
    bottom_right = (int(bottom_right[0]), int(bottom_right[1]))
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0),3)

plt.imshow(img)
plt.show()
```
